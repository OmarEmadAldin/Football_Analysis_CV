```python
from ultralytics import YOLO 
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python

# Define the relative path to the model file
relative_path = r'..\runs\detect\train3\weights\best.pt'

# Normalize the path
best_path = os.path.normpath(relative_path)

# Check if the file exists
if not os.path.exists(best_path):
    raise FileNotFoundError(f"Model file not found: {best_path}")

print(f"Model file found: {best_path}")
```

```python
model = YOLO(best_path)
```

```python
video = (r'E:\Omar 3amora\Football Analysis Proj\Input Videos\Input.mp4')
results = model.predict(video,save=True)
print(results[0])
print('=====================================')
for box in results[0].boxes:
    print(box)
```

```python
results = model.track(source=video ,conf=0.3, iou=0.5, save=True, tracker="bytetrack.yaml")
```
