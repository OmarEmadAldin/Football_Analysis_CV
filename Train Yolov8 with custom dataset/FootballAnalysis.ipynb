# Download Weights
only run one time

```python
import os

if not os.path.exists('yolov8_weights'):
    os.makedirs('yolov8_weights')
```

```python
import requests

# List of YOLOv8 weights URLs
weights_urls = {
    'yolov8n.pt': 'https://github.com/ultralytics/assets/releases/download/v0.0.0/yolov8n.pt',
    'yolov8s.pt': 'https://github.com/ultralytics/assets/releases/download/v0.0.0/yolov8s.pt',
    'yolov8m.pt': 'https://github.com/ultralytics/assets/releases/download/v0.0.0/yolov8m.pt',
    'yolov8l.pt': 'https://github.com/ultralytics/assets/releases/download/v0.0.0/yolov8l.pt',
    'yolov8x.pt': 'https://github.com/ultralytics/assets/releases/download/v0.0.0/yolov8x.pt'
}

# Download each weight and save to the yolov8_weights directory
for weight, url in weights_urls.items():
    response = requests.get(url)
    with open(f'yolov8_weights/{weight}', 'wb') as f:
        f.write(response.content)
```
